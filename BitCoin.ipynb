```python
import pandas as pd
import plotly.offline as py
import plotly.graph_objs as go
from stockstats import StockDataFrame
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
bitcoin=pd.read_excel("Bitcoin.xls", index_col="SNo")
bitcoin["Date"]= bitcoin["Date"].dt.strftime('%d-%m-%Y')

bitcoin
```

```python
#'close ratio' = (Close-Low)/(High-Low)

CloseRatio=pd.DataFrame((bitcoin.High-bitcoin.Low)/(bitcoin.Close-bitcoin.Open))
CloseRatio
```

```python
#'spread'
Spread=pd.DataFrame((bitcoin.High-bitcoin.Low)/(bitcoin.Close))
Spread
```

```python
py.init_notebook_mode(connected=True)
data = [go.Candlestick(x=bitcoin.index,
                       open=bitcoin.Open,
                       high=bitcoin.High,
                       low=bitcoin.Low,
                       close=bitcoin.Close)]
layout = go.Layout(title='Bitcoin Candlestick',
                   xaxis={'rangeslider':{'visible':True}})
fig = go.Figure(data=data,layout=layout)
py.iplot(fig,filename='bitcoin_candlestick')
```

```python
df1=bitcoin
fig = px.scatter(df1, x="Volume", y="Marketcap", trendline="ols")
fig.show()
```
